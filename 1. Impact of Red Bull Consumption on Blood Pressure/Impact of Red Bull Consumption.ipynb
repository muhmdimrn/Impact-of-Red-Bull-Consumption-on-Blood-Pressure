## **Title: "Analyzing the Impact of Red Bull Consumption on Blood Pressure: A Comparative Study Across Gender and Age Groups"**

## Objective:

The primary objective of this analysis is to evaluate the impact of consuming Red Bull on blood pressure among different groups of volunteers. The analysis will focus on the following specific goals:

- Determine the Overall Impact of Red Bull on Blood Pressure
- Assess Gender Differences in Blood Pressure Response
- Investigate Age Group Variations in Blood Pressure Changes
- Identify the Proportion of Volunteers Experiencing Increased Blood Pressure
- Analyze the Relationship Between Initial Blood Pressure and Changes Post-Consumption

## Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the dataset

```python
file_path = 'Redbull.csv'
data = pd.read_csv(file_path)
```

## Explore the data set

Display the top 5 rows

```python
data.head()
```

## Cleaning Data

Remove Unnecessary Columns

```python
data = data.drop(columns=['Unnamed: 5'])
```

Check for Lost Data

```python
data.isnull().sum()
```

## Descriptive Statistical Analysis

### 1. Average Blood Pressure
- Calculate the average of bp_before and bp_after

```python
average_bp_before = data['bp_before'].mean()
average_bp_after = data['bp_after'].mean()

average_bp_before, average_bp_after
```

### 2. Analysis Based on Gender
- Calculate the difference in blood pressure based on gender.

```python
data.groupby('sex')[['bp_before', 'bp_after']].mean()
```

### 3. Analysis Based on Age Group
- Calculate the difference in blood pressure based on age group.

```python
data.groupby('agegrp')[['bp_before', 'bp_after']].mean()
```

### 4. Rate of Volunteers Experiencing Increased Blood Pressure:
- Calculate the percentage of volunteers who experienced an increase in blood pressure.

```python
increase_bp = (data['bp_after'] > data ['bp_before']).mean()*100

increase_bp
```

## Data Visualization

### 1. Visualization of Blood Pressure Changes:
- Create a bar graph to compare the average blood pressure before and after.

```python
# Comparison graph of average blood pressure
sns.barplot(data=data.melt(id_vars=['sex', 'agegrp'],value_vars=['bp_before', 'bp_after']),
            x='variable', y='value', hue='sex')
plt.title('Comparison of Blood Pressure Before and After Red Bull Consumption')
plt.ylabel('Blood Pressure (mmHg)')
plt.xlabel('Condition')
plt.show()
```

### 2. Visualization Based on Age Group:
- Create a box plot to see the variation of blood pressure by age group.

```python
# Plot boxes by age group
sns.boxplot(data=data.melt(id_vars=['agegrp'], value_vars=['bp_before', 'bp_after']),
            x='variable', y='value', hue='agegrp')
plt.title('Blood Pressure Before and After Red Bull by Age Group')
plt.ylabel('Blood Pressure (mmHg)')
plt.xlabel('Condition')
plt.show()
```
